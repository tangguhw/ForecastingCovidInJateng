# Forecasting AR Method

>Model Autoregressive (AR) adalah sebuah model yang digunakan untuk peramalan suatu kejadian dengan menggunakan data-data yang ada pada periode sebelumnya. 

>Bentuk umum model ini adalah: 

>image.png

```python
import pandas as pd
import numpy as np
%matplotlib inline

from statsmodels.tsa.ar_model import AR, ARResults
```

#Load Data Untuk Prediksi

```python
#Load data
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('csv_for_prediction_cases_200516.csv', index_col = 'Tanggal', parse_dates=True)
```

# Grafik Kasus Terkonfirmasi Positif Yang Akan Digunakan Untuk Prediksi

```python
title='Kasus Covid-19 di Jawa Tengah'
ylabel='Jumlah'
xlabel='Tanggal'

ax = df['CentralJava'].plot(figsize=(12,5),title=title);
ax.autoscale(axis='',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

```python
len(df)
```

# Membagi data untuk train dan test

>Data train Digunakan Untuk Melatih Model
*   Data train yang digunakan sebanyak 520

>Data test Digunakan Untuk Mengetest Model
*   Data test yang digunakan adalah 41

```python
train = df.iloc[:520]
test = df.iloc[520:]
train
test
```

```python
import warnings
warnings.filterwarnings("ignore")
```

# Fit an AR(1) Model

```python
model = AR(train['CentralJava'])
AR1fit = model.fit(maxlag=1,method='cmle')
print(f'Lag: {AR1fit.k_ar}')
print(f'Coefficients:\n{AR1fit.params}')
```

```python
start=len(train)
end=len(train)+len(test)-1
predictions1 = AR1fit.predict(start=start, end=end, dynamic=False).rename('AR(1) Predictions')
predictions1
```

```python
#membandingkan prediksi 1 dengan data asli
for i in range(len(predictions1)):
    print(f"predicted={predictions1[i]:<11.10}, expected={test['CentralJava'][i]}")
```

```python
#perbandingan prediksi 1 dengan data asli dalam bentuk grafik
test['CentralJava'].plot(legend=True)
predictions1.plot(legend=True,figsize=(12,6));
```

# Fit an AR(2) Model

```python
AR2fit = model.fit(maxlag=2,method='cmle')
print(f'Lag: {AR2fit.k_ar}')
print(f'Coefficients:\n{AR2fit.params}')
```

```python
start=len(train)
end=len(train)+len(test)-1
predictions2 = AR2fit.predict(start=start, end=end, dynamic=False).rename('AR(2) Predictions')
predictions2 
```

```python
#membandingkan prediksi 2 dengan data asli
for i in range(len(predictions2)):
    print(f"predicted={predictions2[i]:<11.10}, expected={test['CentralJava'][i]}")
```

```python
#perbandingan prediksi 1, prediksi 2, dan data asli dalam bentuk grafik
test['CentralJava'].plot(legend=True)
predictions1.plot(legend=True)
predictions2.plot(legend=True,figsize=(12,6));
```

# Fit an AR(n) model where statsmodels chooses n

```python
ARfit = model.fit(maxlag=30,method='cmle')
print(f'Lag: {ARfit.k_ar}')
print(f'Coefficients:\n{ARfit.params}')
```

```python
start = len(train)
end = len(train)+len(test)-1
rename = f'AR(30) Predictions'

predictions30 = ARfit.predict(start=start,end=end,dynamic=False).rename(rename)
predictions30
```

```python
#membandingkan prediksi 30 dengan data asli
for i in range(len(predictions30)):
    print(f"predicted={predictions30[i]:<11.10}, expected={test['CentralJava'][i]}")
```

```python
#perbandingan prediksi 1, prediksi 2, prediksi 30, dan data asli dalam bentuk grafik
test['CentralJava'].plot(legend=True)
predictions1.plot(legend=True)
predictions2.plot(legend=True)
predictions30.plot(legend=True,figsize=(12,6));
```

# Evaluasi Model

Membantu untuk memiliki sarana perbandingan antara dua atau lebih model. Salah satu metode umum adalah menghitung Mean Squared Error (MSE), tersedia dari scikit-learn.

```python
from sklearn.metrics import mean_squared_error

labels = ['AR(1)','AR(2)','AR(30)']
preds = [predictions1, predictions2, predictions30]  # these are variables, not strings!

for i in range(3):
    error = mean_squared_error(test['CentralJava'], preds[i])
    print(f'{labels[i]} Error: {error:11.10}')
```
